### 3.4. 배열 연산

- 넘파이에서 배열 관련 연산시 반복문의 작성없이 데이터를 처리할 수 있는 기능 제공


- 동일한 크기의 배열간의 연산은 배열의 각각 동일한 원소끼리 연산이 이루어짐

```python
#import numpy as np

arr = np.array([[1,2,3], [1,2,3]])

print(arr.ndim, '\n')
print(arr, '\n')
print('arr * arr = \n', arr*arr, '\n')
print('arr + arr = \n', arr+arr, '\n')
print('arr - arr = \n', arr-arr, '\n')
print('arr / arr = \n', arr/arr, '\n')
print('1 / arr = \n', 1/arr, '\n')
```

#### 다차원 배열의 평탄화 1
- ravel()
    - 원본 데이터에 대한 참조 데이터를 반환함

```python
import numpy as np

arr = np.arange(15)

print(arr, '\n')

arr = arr.reshape(5,-1)

print(arr, '\n')

newArr = arr.ravel()

newArr[0] = 100

print(newArr, '\n')
print(arr, '\n')
```

#### 다차원 배열의 평탄화 2
- flatten()
    - ravel()과 유사하게 동작하지만 원본 데이터의 복사본을 반환함

```python
import numpy as np

arr = np.arange(15)

print(arr, '\n')

arr = arr.reshape(5,-1)

print(arr, '\n')

# TODO: ravel 과 flatten 차이점 => shallow copy, deep copy
newArr = arr.flatten()

newArr[0] = 100

print(newArr, '\n')
print(arr, '\n')

```

### 3.12. meshgrid 함수를 이용한 2차원 그리드 포인트 생성하기

- 변수가 2개인 2차원 함수의 그래프를 그리거나 표를 작성하려면 2차원 영역에 대한 (x,y) 좌표값이 필요


- 그리드 포인트(grid point)를 생성하여 각 좌표에 대한 함수 값을 계산해야 함


- 이러한 그리드 포인트를 만드는 과정을 도와주는 것이 meshgrid 

    
- 가로축의 점들과 세로축의 점을 나타내는 두 벡터를 인수로 받아서 사각형 영역을 이루는 조합을 출력함


- 그리드 포인트의 x 값만을 표시하는 행렬과 y 값만을 표시하는 행렬 두 개로 분리하여 출력함

```python
x = np.arange(3)

y = np.arange(5)

X, Y = np.meshgrid(x,y)

plt.scatter(X,Y)

plt.show()
```

### 공통된 값을 기준으로 해서 두개의 DataFrame 을  병합하는 과정 

### merge() 의 on 옵션을 설정하여 적용합니다

```python
data1 = pd.DataFrame({'key' : ['k0', 'k4', 'k2', 'k3'],
                      'A' : ['A0', 'A1', 'A2', 'A3'],
                      'B' : ['B0', 'B1', 'B2', 'B3']})


data2 = pd.DataFrame({'key' : ['k0', 'k1', 'k2', 'k3'],
                      'C' : ['C0', 'C1', 'C2', 'C3'],
                      'D' : ['D0', 'D1', 'D2', 'D3']})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1, data2, on='key')
```

## 실습 공간

<br>

### merge 과정에서 특정 데이터룰 기준으로 설정할 수 있습니다 

### how 옵션 설정을 합니다.

```python
pd.merge(data1, data2, how='left', on='key')
```

```python
pd.merge(data1, data2, how='right', on='key')
```

## 실습 공간

<br>

### merge 결과를 모두 포함할 수 있습니다. 일종의 합집합과 같은 개념입니다.

### how = 'outer' 옵션을 설정합니다

```python
pd.merge(data1, data2, how='outer', on='key')
```

### merge 결과에서 공통된 요소만 포함할 수 있습니다. 일종의 교집합과 같은 개념입니다.

### how = 'inner' 옵션을 설정합니다

```python
pd.merge(data1, data2, how='inner', on='key')
```
