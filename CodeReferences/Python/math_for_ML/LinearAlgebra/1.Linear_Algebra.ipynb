### 선형대수에서 법선벡터와 한 점이 주어질때 평면 방정식을 정의하는 방법 



<br>

<img src="./images/normalVector.png"  height= 350 width=450 />

<br>

- 법선벡터 즉 한 평면에 직교하는 벡터를 말하여 다음과 같이 정의되어 있다고 가정

$$\vec{n} = (a, b, c)$$

<br>

- xyz 평면위의 임의의 한점을 $(x_p, y_p, z_p)$ 라고 정의하면 위 그림에서 보이는 $\vec{P} - \vec{P_1}$ 는 다음과 같이 정리됨


$$\vec{P} - \vec{P_1} = ( x - x_p,  y - y_p , z - z_p)$$

<br>

- 위 벡터와 법선 벡터는 직교하기 때문에 내적이 0이 됨

$$\vec{n}\cdot(\vec{P} - \vec{P_1}) = 0$$  

<br>

$$a(x-x_p) + b(y-y_p) + c(z-z_p) = 0 $$

<br>

$$ax-ax_p + by-by_p + cz-cz_p = 0 $$

<br>

$$ax + by + cz = ax_p + by_p + cz_p$$

<br>

- 위 식에서 오른쪽 항은 상수값이므로 법선벡터로부터 유도된 평면의 방정식은 다음과 같음

$$ax + by + cz = d$$

- 위 식에 대해서 x, y, z 를 x에 대한 feature 로 간주한다면 다음과 같이 정리할 수 있음 

$$ax_1 + bx_2 + cx_3  -d = 0$$

- 선형대수에서는 일반적으로 열벡터를 가정하고 있기 때문에 위 식을 벡터간의 내적으로 바꿔서 표현

$${
\begin{bmatrix}
a  \\ b \\ c \\
\end{bmatrix}
}^T
\begin{bmatrix}
x_1  \\ x_2 \\ x_3 \\
\end{bmatrix}
- d = 0
$$

<br>

- 위 식은 법선 벡터를 W, x feauture 벡터를 X , 그리고 -d 를 b 라고 하면 최종적으로 다음과 같이 표현할 수 있음

$$W^TX + b = 0$$


- SVM에서 경계면을 초평면이라고도 하는데 이러한 초평면은 위에서 정리한 개념을 바탕으로 구성할 수 있음


- 법선벡터 W가 (2,-4, 7)일 때 어느 한점 (-2, 0,  -4)를 지나는 초평면은 ? 

<br>

$$2x_1 -4x_2 +7x_3 = d$$

$$\Rightarrow \quad d = -32$$

<br>

$$W^TX + 32 = 0$$
<br>

### margin

<img src="./images/svm_margin.png"  height= 350 width=350 />

- 마진의 종류
    - 하드 마진 분류(hard margin classification)
        - 모든 샘플이 바깥쪽 경계면 밖에 존재하는 경우
        - 이상치 데이터가 존재할 경우 일반화하기가 어려움(이상치 민감성 매우 높음)
    - 소프트 마진 분류(soft margin classification)
        - 일부 데이터가 섞여 있는, 어느 정도의 마진 오류 허용
        - 하드 마진 분류에 비해 유연하게 일반화를 이룰수 있음
        
        
- SVM 역시 범주형 모델이기 때문에 위 그림에서 보이듯이 데이터들을 일정한 기준에 맞추어 두 범주로 분류하고 있음


- 두 범주로 나눌 수 있는 분류 기준은 위 그림에서 알 수 있듯이 두 범주로 구분할 수만 있다면 $H_1$ 과 $H_2$ 와 같은 여러 형태가 있을 수 있음


- 예시에서 보이는 구분선 $H_1$ 와 $H_2$ 모두 두 클래스를 무난하게 분류하고 있음을 확인할 수 있음 


- 좀 더 나은 분류 경계면은 $H_2$ 이라고 할 수 있음 


- $H_1$ 에 비해  두 범주를 여유있게 가르고 구분하고 있기 때문


- 예시에서 보이듯이 구분선 $H_2$ 을 기준선으로 가정
    - Supporting Vector 사이의 거리를 마진(margin)이라고 함


- 따라서 마진폭이 크면 클수록 데이터를 분류하는데 있어서 오류를 줄일수 있는 보다 편리하고 안전한 방법을 제공함


- SVM은 이 마진을 최대화하는 분류 경계면을 찾는 알고리즘


- $H_2$ 에 해당하는 경계면, 즉 이 경계면이 분류의 기준이 되는 결정 경계면(decision boundary)이라고 함

### 선형 SVM 분류 

- 이 결정 경계면을  $W^TX + b = 0$ 라고 하고 가정


- 예를 들어 흰 원이 나타나는 곳을 +1 , 검은 원이 나타나는 곳은  -1 라고 가정함 
    - 즉 위 방정식의 값은 두가지만 존재한다고 가정 => 분류의 문제이기 때문


<br>

<img src="./images/svm_plane.png"  height= 350 width=350 />

<br>

#### 예시에서 $X^+$ 평면과 $X^-$ 평면의 거리를 최대화할수 있도록 해야  하기 때문에 두 평면간의 거리를 구하는 과정은 다음과 같음

- 예시에서 $X^+$ 평면과 $X^-$ 평면의 거리를 최대화할수 있도록 해야   
하기 때문에 두 평면간의 거리를 구하는 과정은 다음과 같음


- 거리를 구하고자 하는 두 평면의 각 점을 다음과 같이 정의

<br>

$$||X^p - X^n||\;\;(\; X^p : plus-plane 상의\;한\;점, X^n : minus-plane\;상의\;한\;점)$$

<br>

- 또한 각 점은 각 경계면상의 점들이기 때문에 다음과 같이 정리할 수 있음



- $||X^p - X^n||$ 을 K 라고 가정하고 $X^p$ 와 $X^n$ 은 각각의 평면에 있는 벡터임


- $X^p - X^n$  은 법선 벡터 W 을 가지고 크기가 1인 단위 벡터로 만들면 다음과     
같이 정리할 수 있음

$$X^p - X^n\;\;=\;\;K * \frac{W}{||W||}$$ 


$W^TX^P + b = 1$  => (1)

$W^TX^ n + b = -1$ => (2)

<br>

위 두식에서  (1) - (2) 연산을 하면 다음과 같습니다.

<br>

$W^T(X^P - X^ n)  = 2$  이 식에 위에서 법선 벡터를 가지고 정리한 식을 대입하도록 합니다.

<br>

$\displaystyle W^T(K * \frac{W}{||W||})  = 2$  

<br>

$\displaystyle K * \frac{W^T\;W}{||W||}  = 2$    ( $W^TW =  (||W||)^2$  )

<br>

$K * ||W||  = 2$  이 되므로 margin 에 해당하는 거리 K 는 다음과 같습니다.

<br>

$$K  = \frac{2}{||W||}$$

<br>

- margin $\displaystyle K  = \frac{2}{||W||}$ 을 최대화하여  모든 데이터들에 대해서 $W^TX^P + b \ge 1 $ 을 만족시키거나   
   $W^TX^ n + b \le -1$ 을 만족시켜서 분류를 되도록이면 오차없이 할 수 있도록 해야함



<br>

- 즉 데이타의 분류 결과값 y == 1 이면 $W^TX^P + b \ge 1 $, y == -1 이면  $W^TX^ n + b \le -1$ 이므로   
  최종적으로 하나의 식으로 나타내어보면 다음과 같음. 

<br>

$$y*(W^TX + b ) \ge 1$$

<br>

- margin 을 최대화하기 위해서는 $K  =\displaystyle \frac{2}{||W||}$ 을 최대화시켜야 함


- 이 식의 의미는 $\frac{||W||}{2}$ 을 최소화시켜야 한다는 것과 동일한 개념


- 최소화시키는 대상인 $\displaystyle \frac{||W||}{2}$ 에 대해서는 일반적으로 제곱값을 최소화시키는 $\displaystyle \frac{\;(||W||)^2\;}{2}$ 으로 정함


- 최적화의 관점에서 고려해볼때  찾고자하는 W 는 $y*(W^TX + b ) \ge 1$ 을 만족시키는 조건하에   
   $\displaystyle \frac{\;(||W||)^2\;}{2}$ 을 최소화하는  W를 찾는 것

### scikit-learn 을 이용한 SVM 구현 1 

#### 1. 데이터 로딩

```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris.data.shape)
print(iris.feature_names)
```

### 실습공간

```python
# iris feature 에서 2개만 사용합니다.
#X = iris.data[:, [2,3] ] 
X = iris.data[:, :2] 
y = iris.target
```

### 실습공간

#### 2. training data / test data 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)
```

### 실습공간

#### 3. 데이터 scaling

- StandardScaler
    - 데이터의 특성의 평균을 0, 분산을 1로 변경함 
    - 데이터의 특성의 최소값과 최대값에 대한 제한을 가하지는 않음 
    
    
- SVM 은 특성의 스케일에 민감한 머신러닝
    - 특성의 스케일을 조절하면 결정 경계가 훨씬 좋아짐
    
<br>

<img src="./images/svm_Scale1.png"  width=650 />

<br>

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train) 
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

### 실습공간

#### 4. SVM training

- C 
    - 기본값으로 1.0 설정
    - 얼마만큼의 여유를 가지고 오류를 인정할건지에 대한 값
    - C 값을 높이면 마진폭이 좁아지며 SVM이 더 잘 분류하여 마진 오류 적게 발생
        - 그러나 커질수록 과대 적합 가능성이 커짐
    - C 값을 낮추면 마진폭이 넓어지지만 마진 오류도 커짐 

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train_std, y_train)
```

### 실습공간

#### 5. SVM 성능 측정

```python
y_pred = svm.predict(X_test_std)

print("MisClassified Examples : %d" % (y_test != y_pred).sum())

from sklearn.metrics import accuracy_score
print("Accuracy : %.4f" %accuracy_score(y_test, y_pred))
```

### scikit-learn 을 이용한 SVM 구현 2

```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris.data.shape)
print(iris.feature_names)
```

### 실습공간

<br>

```python
# iris feature 에서 2개만 사용합니다.
#X = iris.data[:, [2,3] ] 
X = iris.data[:, :2] 
y = iris.target
```

### 실습공간

<br>

-  hinge loss ?
    - 마진을 극대화하는 SVM에서 사용하는 손실함수
    
    $l(y) = max(0, 1 - t*y)$
    
    - t : 예측 결과
    - y : 원래 타겟 결과 
    - t 와 y가 동일한 sign 값을 가진다면 즉 , 분류결과가 정답이면 손실함수는 0이됨
    - t 와 y가 다른다면 손실 증가

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)

linearSVM = Pipeline([
                ('scaler', StandardScaler()),
                ('linear_svc', LinearSVC(C=1, loss='hinge')) ])

linearSVM.fit(X_train, y_train)

y_pred = linearSVM.predict(X_test)

print("MisClassified Examples : %d" % (y_test != y_pred).sum())

from sklearn.metrics import accuracy_score
print("Accuracy : %.4f" %accuracy_score(y_test, y_pred))
```

#### KNN 알고리즘의 처리 과정


1. k 에 해당하는 숫자와 거리 메트릭을 선택합니다.  

2. 분류하고자 하는 데이터에 대한 k 개의 근접한 이웃을 찾습니다.  

3. 다수결 방식에 의해서 데이터의 분류 레이블을 결정합니다.   

4. 거리를 측정할때는 유클리드 거리 측정법을 사용

다음 그림을 보면 분류하고자 하는 데이터가 물음표(?)로 표시되어 있으며 주위에 근접한 다섯개의 근접한 데이터를 기준으로 다수결 원칙에 의해서 삼각형 레이블로 분류되는 과정을 보여주고 있습니다.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import numpy as np
        
iris = datasets.load_iris()

# iris feature 에서 2개만 사용합니다.
X = iris.data[:,[2,3]]  
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)
sc = StandardScaler()
```

```python
sc.fit(X_train) 
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski') # p = 2 & 'minkowski' 두 옵션이 같이 쓰여야 유클리디안 거리 옵션이 적용됨
knn.fit(X_train_std, y_train)

y_pred = knn.predict(X_test_std)

from sklearn.metrics import accuracy_score
print('MisClassification Exampes ; %d' % (y_test != y_pred).sum())
print('Accuracy : %.2f' % accuracy_score(y_test , y_pred))
```

   데이터를 이용해 여러개의 객체를 모아 집단으로 구분하는 작업


- 라벨이 없는 데이터 각각의 특성을 고려하여 주어진 데이터들을 가장 설명하는    
   몇 개의 집단으로 나누는 것을 의미


- 대표적 알고리즘인 K-means는 데이터의 K개의 클러스터로 구분하는    
   알고리즘이며 K값에 따라 달라짐


- K개의 클러스터로 나누는 기준은 데이터들간의 유사도를 가지고 판단하며    
   이는 데이터간의 거리를 측정하여 계산함


### K-means 알고리즘 원리

- 클러스터내 데이터간 거리합 최소화
    - 클러스터내 데이터와 클러스터 중심점 간의 거리 최소화 
   

- 클러스터간 분리도 최대화
    - 클러스터간의 중심점 거리 최대화


- 거리 측정법
    - 기본적으로 유클리드 거리 측정법을 이용
    - 벡터간의 거리가 짧으면 유사성이 높다는 의미
    - 중심에서 멀리 떨어져 있어도 K-means 알고리즘은 반복적으로    
       중심점 수정할 수 있음 
    

### K-means 알고리즘 수행 과정

- 전체 클러스터 개수 K를 설정 


- 초기 중심점 설정
    - 임의의 K개의 점을 선택하여 K개의 클러스터 초기 중심점 설정
    
    
- 클러스터 할당
    - K개의 클러스터 중심점과 각 데이터간의 거리를 측정하여 가장     
       가까운 클러스터 할당
  
  
- 새로운 중심점 설정
    - 각 클러스터마다 그안에 배정된 모든 점들간의 거리 평균값을    
       구하여 새로운 중심점 설정
    

### K-means 테스트

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs # make_blobs : 클러스터링 모델 시험시 원하는 특성을 가진 데이터 만들기 위해 사용함

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

centers = [[1,1],[0,0],[2,-1]]
n_clusters = len(centers)

data, labels_true = make_blobs(n_samples = 2000, centers=centers, cluster_std = 0.7)

print(data)

print(labels_true)

print(np.unique(labels_true)) # 생성된 데이터들이 0,1,2 클러스터에 속하는 것을 확인할 수 있음
```

- 현재 생성된 데이터들이 분포 현황을 알기 위해 산포도를 통해 확인

```python
plt.figure(figsize=(10,5))

plt.scatter(data[:,0], data[:,1])
```

- n_init : 초기 중심점 선택시 최상의 결과를 얻기 위한 시도 횟수

```python
estimator = KMeans(init='k-means++', n_clusters=3, n_init=10)
estimator.fit(data)
```

- 학습후 클러스터 라벨값 확인

```python
labels_pred = estimator.labels_
labels_pred
```

```python
cm = plt.cm.get_cmap('jet')
scaled_labels = labels_pred - np.min(labels_pred)
scaled_labels = scaled_labels/(np.max(labels_pred) - np.min(labels_pred))
np.unique(scaled_labels)
```

```python
plt.figure(figsize=(10,5))

plt.scatter(data[:,0], data[:,1], c=cm(scaled_labels))
```

# 5.1. 역행렬
- 역행렬은 교환법칙이 성립되며 항등행렬을 만들어주는 역할을 수행  

$$A_{n \times n}A^{-1} = A^{-1}A_{n \times n} = I$$ 


- 모든 정방행렬이 역행렬을 가지는 것은 아님 


- 다음 조건을 만족할 때 역행렬 존재  
    - $n \times n$ 행렬일때 $n$개의 pivot이 만들어졌을때  
        - 가우스 소거시 대각선에 0이 아닌 값만 존재하는 경우 

    - $\det{A} \neq 0$
        - $\det{A} = \prod_{i = 1}^nd_i$  

    - 역행렬은 유일하게 하나만 존재함 


- 역행렬이 존재하면 유일한 해을 구할수 있는 조건이 만족됨
    - $Ax = b$ 

      $A^{-1}Ax = A^{-1}b$ 

      $Ix = A^{-1}b$
      
 <img src="./images/inversesystem.png" width=300>
 <br>    

- 0이 아닌 벡터 $x$ 가 다음 방정식을 만족시킬때 역행렬은 존재할수 없음
    - $Ax = 0$ 

        역행렬이 존재한다면 0 이 아닌 벡터라는 전제가 성립이 안됨
        -  $A^{-1}Ax = 0$ 

           $\rightarrow \quad Ix = 0$


- $2 \times 2$행렬의 경우 역행렬은 다음과 같은 공식에 의해 유도됨

 $$A = \begin{bmatrix} 
 a & b  \\\\

 \end{bmatrix}
 \rightarrow A^{-1} = \frac{1}{ad-bc}
 \begin{bmatrix} 
 d & -b  \\\\ 
 -c & a  \\ 
 \end{bmatrix} \quad ( ad-bc : \det)
 $$ 

<br> 

- 3차원 이상의 행렬에 대해서 역행렬을 구하는 절차는 가우스 조단을 통해 구함

 \end{bmatrix}
 $$

#  5.3. 전치행렬  
- $A^T$ 로 표현  


- 전치 : $a_{ij}  \rightarrow a_{ji}$


- $(AB)^T = B^TA^T$  


- $(A+B)^T = A^T + B^T$ 


- $(A^{-1})^T = (A^T)^{-1}$ 


- 대칭행렬(Symmetric Matrix)
    - $A^T = A$ (행렬이 정방행렬일때 성립)
    - $a_{ij} = a_{ji}$
    - 역행렬을 가지고 있다면 $A^{-1}$ 역시 대칭행렬임  
         - $(A^{-1})^T = A^{-1}$

    - $A = LU = LDU$
        - $A^T = A$

        - $LDU = U^TDL^T$  
            - $L = U^T$
            - $U = L^T$

# 6.2. 벡터 공간 (Vector Space)
- 임의의 벡터 $\vec{X},\vec{Y} \in \mathbb{R^n}$ , 임의의 스칼라 $c \in \mathbb{R}$ 일때 다음 조건을 만족하는 집합을 벡터 공간이라고 함 

    - $\vec{X}+\vec{Y} \in \mathbb{V}$
    - $c\vec{X} \in \mathbb{V}$
    - $c\vec{X}+c\vec{Y} \in \mathbb{V}$


-  벡터간에는 교환 법칙 성립
    - $\vec{X}+\vec{Y} = \vec{Y}+\vec{X}$ 


- 벡터간에는 분배 법칙 성립
    - $c(\vec{X}+\vec{Y}) = c\vec{X}+c\vec{Y}$ 


- 벡터간에는 결합 법칙 성립 
    - $\vec{X}+(\vec{Y}+\vec{Z}) = (\vec{X}+\vec{Y})+\vec{Z}$ 


- 벡터 공간은 항등원 벡터 $\vec{0}$를 가지고 있음  
    - $\vec{X}+ \mathbb{O} = \mathbb{O} +\vec{X}= \vec{X}$   


- 벡터 $\vec{X}$는 유일한 $-\vec{X}$ 를 가지고 있음
    - 앞서 정의한 항등원 벡터가 존재한다는 전제  

    - $\vec{X} + (-\vec{X})  = \vec{0}$


- 행렬은 벡터의 집합이므로 행렬에 대해서 벡터공간을 정의할수 있음

# 7.2. column space of 행렬 A
- 행렬 A에 있는 모든 column vector의 선형 결합으로 이루어진 공간  


$$
 \begin{bmatrix} 
 \\\\ 
 a_1 & a_2 & \cdots & a_n \\\\ 
 \\ 
 \end{bmatrix}
$$  

<br>  

- 행렬 A가 위와 같은 컬럼 벡터로 표현될 때 다음과 같은 column space를 정의할 수 있음  

<br>

$$\mathbb{C}(A) = \{{X:X=\sum_{i=1}^{n}c_ia_i}\}$$  

<br>

- $Ax=b$ 라는 방정식은 다음과 같은 형식으로 표현됨

$$
 \begin{bmatrix} 
 \\\\ 
 a_1 & a_2 & \cdots & a_n \\\\ 
 \\ 
 \end{bmatrix}
 \begin{bmatrix} 
 x_1\\\\ 
 x_2\\\\ 
 \vdots\\\\ 
 x_n
 \end{bmatrix}
 =
 b
$$  

<br>

$$x_1a_1 + x_2a_2 + \cdots + x_na_n = b$$

<br>

- 위 식에서 보이듯이 $b$ 라는 벡터는 column space의 선형결합과 같음을 확인할 수 있음   


- $if \; b \in \mathbb{C}(A) \rightarrow 해집합 \; 존재$
    - 물론 해가 특정 해인지 무수히 많은 해인지는 추후에 살펴봐야 함  


- $if \; b \notin \mathbb{C}(A) \rightarrow 해집합 \; 존재하지 \;않음$  

<br>

- $Ax=b$ 에서 $A^{-1}$ 가 반드시 존재하면 다음 식이 성립됨  
    - 행렬 $A$ 는 $n \times n$ 정방행렬
    - $x=A^{-1}b$ 
    - 즉 유일한 해가 존재한다는 의미가 됨
    - $A^{-1}$ 존재하면 $b$는 항상 column space내에 존재한다는 의미

    - $cx \in \mathbb{N}(A)$  

    - $A(cx) = cAx = \mathbb{O}$

# 8.1. 선형 독립(linear independence)   

- 행렬의 크기 $m \times n$은 연립 일차 방정식의 참된  크기에 대한 완전한 정보를 제공하지 못함.

- $c_1V_1 + c_2V_2 + \cdots + c_nV_n = 0$ 일때
    - 오직 $c_1 = c_2 = \cdots = c_n =0$ 이 영 벡터를 구성하는 유일한 방법일 때 벡터들간에 선형 독립 만족

# 10.1. 벡터 직교의 잇점
- 벡터가 서로 수직(orthogonal)이면 ?  

    - 벡터는 서로 선형 독립임(linear independent) 
  
    - 기저벡터를 구성할 때 직교인 벡터로 구성하면 많은 잇점이 있음  
    
        - $\sum_iC_iV_i = X$
    
        - 특정 i 번째 벡터와 내적을 구하면 다음과 같고 그럼 선형 결합을 구성하는 계수를 쉽게 구할수 있음 
    
        -  $X^TV_i = C_i{||V_i||}^2$
     
    - 벡터 연산이 단순하고 쉽게 처리가능함

# 10.2. 벡터의 직교
- 벡터의 길이 : ||X|| 

    - ${||X||}^2 = X^TX = \sum_{i=1}^nx_i^2$ 




- 벡터 $X, Y$가 있을 때 내적 $X^TY$
    - $X^TY = 0 \quad \rightarrow$  벡터는 서로 수직


- 0이 아닌 벡터 $V_1, v_2, \cdots, V_n$가 서로 직교하면 다음과 같은 성질 만족 

    - $V_i^TV_j = 0 \quad ( i \neq j)$   

    - 벡터들은 서로 linearly independent  
        - $C_1V_1+ C_2V_2+ \cdots + C_nV_n = 0 \quad(C_1=C_2=\cdots=C_n = 0)$   
        $V_i^T(C_1V_1+ C_2V_2+ \cdots + C_nV_n) = 0$  
        $C_i{||V_i||}^2 = 0 ,\quad \forall i$  
        $C_i = 0 ,\quad \forall i$  


- 벡터 공간의 직교
    - 한 SubSpace에 있는 모든 벡터가 다른 SubSpace에 있는 모든 벡터와 수직일 때  

    - $\mathbb{C}(A^T) \perp \mathbb{N}(A)$  

    - $\mathbb{C}(A) \perp \mathbb{N}(A^T)$

# 12.1. Gram-Schmidt
- 선형 독립인 벡터들이 있을 때 정규 직교인 기저 벡터로 표현하는 방법 정의  


- 어떤 임의의 행렬 A의 열벡터(column vector)를 정규 직교(orthonormal) 벡터로 변환하는 과정임 


- 서로 정규 직교하는 벡터로 이루어진 행렬 A 직교 행렬(orthogonal matrix) 이라고 함   


- 위에서 b를  앞서 다뤘던 투영에 대한 방법을 이용하여 a에 투영을 한다.<br>
    - 먼저 벡터 a를 정규화시켜 $q_1$을 정의한다.   
        - $q_1 =\displaystyle \frac{a}{||a||}$<br><br>
    
    - b를 투영시키면 다음과 같다. 
        - $\displaystyle \frac{q_1^Tb}{q_1^Tq_1}q_1 = q_1^Tbq_1, \quad\quad q_1 \Rightarrow ||q_1|| = 1$  


<br>

<br> 

- b에서 a에  수직인 벡터를 이용하여 $q_2$를 구한다. 
    - $q_2 =\displaystyle \frac{b-q_1^Tbq_1}{||b-q_1^Tbq_1||}$<br><br>   
    
    - b = $q_1^Tbq_1 + q_2^Tbq_2$

- 이제는  $q_1, q_2$에  수직인 $q_3$를 구해보자 
    - $q_3 =\displaystyle \frac{ c - (q_1^Tcq_1 + q_2^Tcq_2)}{||c - (q_1^Tcq_1 + q_2^Tcq_2)||}$<br><br>  
    
    - c = $q_1^Tcq_1 + q_2^Tq_2 + q_3^Tq_3$

    - $a_2-({q_1}^Ta_2)q_1 \; \perp \; q_1$ <br><br>

    - $q_j \rightarrow a_j - \sum_{i=1}^{i-1}({q_i^Ta_j})q_i$ 정규화<br><br>   
    - $a_j = \sum_{i=1}^j({q_i^Ta_j})q_i$

# 13.1.  고유값 / 고유벡터  
- $Ax = \lambda x$
    - $A$ : 정방행렬 

    - $\lambda$ : scalar, 고유값(eigen value)

    - $x$ : 고유벡터(eigen vector) 

<br> 

$$Ax = \lambda x$$

<br>   


$$(A- \lambda I)x= 0$$

- 위 식에서 $A- \lambda I$ 는 역행렬이 존재하면 $x$ 는 항상 0 됨    

    - 따라서 $A- \lambda I$ 는 역행렬이 존재하면 안됨 

    - $\det{(A- \lambda I)} = 0$ 을 만족해야 함  


- 행렬 A가 대칭 행렬이면 다음과 같은 특징이 있음  
    - 서로 다른 고유값들에 대응하는 고유 벡터들은 서로 직교함
    - 고유 벡터들을 정규화하면 대칭행렬의 고유 벡터 행렬은 정규직교 행렬이 됨   
    - 정규 직교 행렬은 역행렬이 전치행렬과 같음   
    

- 고유값/고유벡터 관련 예

$$A=\begin{bmatrix} 4 & -5 \\\\  2 & -3 \\ \end{bmatrix}$$

<br><br> 

$$\det{(A- \lambda I)} =
\begin{vmatrix} 4-\lambda & -5 \\\\  2 & -3-\lambda \\ \end{vmatrix}
= (4-\lambda)(-3-\lambda) + 10  = 0
$$  

<br><br>

$$\lambda = 2, \; -1$$

<br><br>

$$\lambda = 2, \quad\quad 
A=\begin{bmatrix} 2 & -5 \\\\  2 & -5 \\ \end{bmatrix}$$

<br>  

- 현재 고유값을 구했으며 그 고유값을 가지고 행렬을 구할 수 있음   

<br>  

$$ 
\begin{bmatrix} 2 & -5 \\\\  2 & -5 \\ \end{bmatrix}
\begin{bmatrix} x_1 \\\\  x_2 \\ \end{bmatrix}
 = 0
$$

<br>  

- 행렬식을 대해 Null Space 해를 구하는 과정이 고유 벡터를 구하는 것임   


- $\lambda = 2$일때 고유 벡터 구하기   

$$ 
\begin{bmatrix} 2 & -5 \\\\  2 & -5 \\ \end{bmatrix}
\begin{bmatrix} x_1 \\\\  x_2 \\ \end{bmatrix}
 = 0
 , \quad \Rightarrow
\begin{bmatrix} \displaystyle \frac{5}{2} \\\\  1 \\ \end{bmatrix}   
\Rightarrow
\begin{bmatrix} 5 \\\\  2 \\ \end{bmatrix} 
$$



- $\lambda = -1$일때 고유 벡터 구하기

- $AX = \lambda X$ 이고 $X$ 가 고유벡터, $\lambda$ 가 고유값이라면 다음과 같음  
<br> 

$$

$$A

# 13.3. 머신러닝 이해하기 - 비지도학습 PCA를 이용한 차원 축소

### PCA 알고리즘 적용과정

1. N개의 데이터를 각각 벡터로 표현 


2. 벡터를 행렬로 구성


3. 각 열로부터 평균값을 빼서 중심으로 이동시키고 B 행렬을 다시 구성


4. 공분산 행렬 C = $\displaystyle \frac{1}{N}B^TB$ 계산


5. C 행렬의 고유값/고유벡터를 구함


6. 가장 작은 고유값부터 제거하여 차원을 축소함


### PCA 세부 사항

- 고차원의 데이터를 저차원의 데이터로 변환하는 차원 축소 방법


- PCA를 사용해 찾은 주성분은 데이터 분포의 특성을 가장 잘 설명해 줄 수 있는 벡터


- 즉 데이터의 분산이 가장 큰 축이 첫번째 주성분이 되고 두번째로 분산이 큰 축이 두번째 주성분이 됨


- PCA 를 통해 사용해 찾은 요소들은 데이터의 분포를 가장 잘 나타내고 있는 요소들이므로 이를 사용해 데이터를 분석할수 있음


- 주성분은 고유벡터와 고유값을 사용해 구함


- 고유벡터와 고유값은 항상 쌍으로 존재하며 이때 고유 벡터는 데이터의 분포를 나타내는 선


- 고유값은 해당 고유 벡터에 대해 데이터를 사상시켰을 때 데이터가 분포하는 분산을 의미

 
- 고유값이 크면 큰 변화량을 가지고 있다는 의미


- 고유값이 작으면 대부분의 데이터들이 밀접하게 붙어있어서 그 방향으로 큰 변화량이 없다는 의미
  


#### 이렇게 데이터의 주성분을 구해서 데이터에 어떻게 적용해야 하는지 알아보겠습니다. 

- 우선 데이터의 고유벡터와 고유값 쌍의 개수는 데이터가 가지고 있는 속성 , 차원의 개수와 같습니다. 


- 예를 들어 데이터가 2개의 속성을 가지고 있을 경우 이 데이터의 고유벡터와 고유값 또한 2쌍입니다. 


- 따라서 2차원 데이터일 경우 2개의 고유 벡터가 수직일 때 데이터가 두 고유벡터를 기준으로 큰 분산으로 분포하고 있다는 것을 의미합니다.  


- 그러면 이 두 고유벡터를 기준으로 데이터를 재구성하는 것이 가능하게 됩니다. 이 때 데이터에는 변화가 생기지는 않고 다만 데이터의 분포를 나타내는 축만 변화하게 됩니다

# 15. 특이값 분할(SVD)
## **학습개요**
- 특이값 분할

<br> 

# 15.1. 특이값 분할(SVD)
- 고유값/고유벡터는 정방행렬일 경우에만 행렬 대각화 및 행렬분할 가능  


-  특이값 분할은 직사각형 행렬일때 행렬을 대각화하고 분할하는 방법임 


- $A = U\Sigma V^T$  
    - 행렬 $A$는 $m \times n$

    - $U$는 $m \times m$   

    - $\Sigma$는 특이값들을 대각요소로 갖는 행렬 $m \times n$ 

    - $V$는 $n \times n$  

    - $\mathbb{C}(A) \perp \mathbb{N}(A^T)$   

    - $\mathbb{C}(A^T) \perp \mathbb{N}(A)$   
    

- 행렬 A가 $m \times n \;$이라면 $A^TA \;$는 다음과 같은 특징을 갖음 
    - $A \;$  와  $A^TA \;$는 Null Space 가 같음  
    
    - $A \;$  와  $A^TA \;$는 Row Space 가 같음  
    
    - $A \;$  와  $A^TA \;$는 Column Space 가 같음  
    
    - $A^TA \;$는 대칭행렬이 되며 대칭 행렬일 경우 서로 다른  고유값에 대응하는 고유 벡터들은 서로 직교함
       
  
- SVD 구하는 과정 
    - 행렬 $A(m \times n)$ 에 대해서가 아닌 행렬 $A^TA$ 고유값/고유벡터를 구함   
        - $A^TA$ : 대칭행렬, $n \times n$
        - 고유값은 0보다 크거나 같은 성질이 있음   
            
    - $r$개의 0이 아닌 고유값이 있다고 가정하자       
         - $\lambda_1, \lambda_2, \cdots, \lambda_r$  : 0 이 아닌 고유값
        
         - $\lambda_{r+1} = \cdots = \lambda_n = 0$   

         - 행렬 $A$ 특이값은 0이 아닌 고유값에 루트를 적용한 값  

         - 특이값 $\sigma_i = \sqrt[]{\lambda_i}, \; (i = 1,2, \cdots, r)$ 

         - 각각의 고유값에 해당하는 고유벡터를 구함 

         - 고유벡터를  정규 직교하는 벡터로 변환시킴 (orthonormal)

         - $A^TA$가 $n \times n$이므로 고유벡터는 $R^n$에 속함
    
<br> 

$$
\Sigma_{1, r \times r} =
\begin{bmatrix} 
\sigma_1 & 0 &\cdots & 0 \\\\
0 & \sigma_2 & \cdots & 0 \\\\
\vdots \\\\
0 & 0 & \cdots & \sigma_r \\
\end{bmatrix} 
\quad \Rightarrow
\Sigma_{m \times n} =
\begin{bmatrix} 
\sum_{1, r \times r} & 0  \\\\
0 & 0 \\
\end{bmatrix} 
$$

<br><br> 


- $V = [v_1, v_2, \cdots, v_r] \in R^n$  

    - $A^TA$가 $n \times n$이므로 고유벡터는 $R^n$에 속함

    - 행렬 $A$를 기준으로 살펴보면 $\mathbb{C}(A^T)$ Row Space를 구성하는 기저벡터로서 사용될 수 있음

    - $\lambda_{r+1} = \cdots = \lambda_n = 0 \;$ 를 살펴보자 
        - 현재 $A^TA$에 대한 고유값임

        - $A^TAx_j = \lambda_jx_j = 0$ 

        - $x_j \Rightarrow \mathbb{N}(A^TA) = \mathbb{N}(A)$

        - $x_j$는 $\mathbb{N}(A)$를 구성하는 기저벡터로 사용할 수 있음 

    - $v_{r+1}, \cdots, v_n \Rightarrow \mathbb{N}(A)$ 기저벡터 

    - $V_1 = [v_1, v_2, \cdots, v_r]$ : $\mathbb{C}(A^T)$ 기저벡터

    - $V_2 = [v_{r+1}, v_{r+2}, \cdots, v_r]$ : $\mathbb{N}(A)$ 기저벡터 

    - $V=[V_1, V_2]$
    
    - $V$는 다음과 같은 특성을 가지고 있음  
        - $V^TV = I$


- $A = U\Sigma V^T$  

     - $AV = U\Sigma$

     - $AV = A[V_1, V_2, \cdots, v_r, V_{r+1}, \cdots, V_n]$  
    
     - $AV = [AV_1, AV_2, \cdots, AV_r, 0, \cdots, 0]$  

     - $U = [u_1, u_2, \cdots, u_m]$ : $m \times m$
    
     - $\Sigma$는 특이값을 갖는 대칭행렬 : $m \times n$

     - $U \Sigma = [\sigma_1 u_1, \sigma_2 u_2, \cdots, \sigma_r u_r, 0, \cdots, 0]$ : $m \times n$ 

     - $AV_i = \sigma_iu_i$  
        - $u_i =\displaystyle  \frac{1}{\sigma_i}AV_i, \; \in R^m, \; (i = 1, 2, \cdots, r)$  
        
        - $u_i^Tu_j =\displaystyle  \frac{1}{\sigma_i \sigma_j}{V_i}^TA^TAV_j = \frac{1}{\sigma_i \sigma_j}{V_i}^T \lambda_jV_j$    
            - $i, \; j \;$가 서로 다르면 위 식은  0 이 됨  

        - $u_i$ 는 정규 직교 기저벡터의 특징을 가짐 

        - $\mathbb{C}(A)$ 기저벡터로서 사용할 수 있음   

    - $U_1 = [u_1, u_2, \cdots, u_r]$ : $\mathbb{C}(A)$ 기저벡터 

    - $u_{r+1}, \cdots, u_m \in R^m$  

        - left Null Space : $\mathbb{N}(A^T)$ 기저 벡터   

        - $U_2 = [u_{r+1}, \cdots, u_m]$  

    - $U = [U_1, U_2]$
