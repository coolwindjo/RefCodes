# 1.3. 확률의 정의
- 공리적 정의 
    - 표본 공간을 구성하는 각각의 사건 $A_i$ 는 0 과 1 사이의 수로 표현됨  
        - $0 \le P(A_i) \le 1$ 

    - 특정 사건에 대한 확률의 모든 합은 1  
        - $\sum_iP(A_i) = 1$  

    - 서로 배반인 사건 $A_1, A_2, \cdots, A_n$
        - $P(A_1 \cup A_2 \cup \cdots = \sum_{i=1}^nP(A_i)$  


- 상대 빈도 정의
    - 랜덤 실험이 n번 시행되었다고 했을 때 사건 A가 $n_A$ 번 일어났다면 확률은 다음과 같이 정의됨   

    - 확률에 대한 상대빈도의 정의는 직관적으로 이해할수 있지만 실제로 실험에 대한 무제한 반복은 있을 수가 없음  

   $$P(A) = \displaystyle lim_{n \to \infty}\frac{n_A}{n}$$  


- 고전적 정의  

    - 사건 A의 확률은 실험의 가능한 전체 결과의 수 N에 대한 특정 사건 A에 해당하는 실험의 결과의 수 $N_A$의 비율로 정의함  

    - $P(A) = \frac{N_A}{N}$  

    - 이 확률은 실험을 실제적으로 수행하기 전에 이미 결정됨   

        - priori 
    
        - 동전던지기는 앞면이 나올 확률을 미리 알 수 있음

# 1.5. 확률의 법칙 
- 표본 공간을 구성하는 각각의 사건 $A_i$ 는 0 과 1 사이의 수로 표현됨  
  $0 \le P(A_i) \le 1$ 


- 특정 사건에 대한 확률의 모든 합은 1  
  $\sum_iP(A_i) = 1$  


- 상호 배타적인 사건 $A_1, A_2,A_3,\cdots$에 대해서  
    - $P(A_1 \cup A_2 \cup A_3 \cdots) = P(A_1) + P(A_2) + \cdots$    


- 배반 사건(disjoint event)   
    -동시에 일어날 수 없는 사건, 두 사건의 교집합이 공집합이 되는 사건 


- 사건 A의 여사건 : 표본 공간에서 A를 제외한 사건을 말함 : $A^c$


- 사건 A의 여사건이 발생할 확률 

    - $P(A^c) = 1 - P(A)$  

    - 1000장중에 4장이 당첨복권, 4장을 구입한다면 적어도 한장 이상의 당첨복권을 구입하게 될 확률은 ?

        <br> 

        - $P(A^c) =\displaystyle \frac{996*995*994*993}{1000*999*998*997} = 0.9841$

$$P(A|B) = \frac{P(A\cap B)}{P(B)}$$

    - 확률 실험에서 새로운 정보 또는 조건(A)이 추가되었을 때 사건 B의 확률

예제) 조건부 확률 1

가방안에 8개 빨간공, 4개 녹색공, 8개 노란공이 있음, 하나의 공을 꺼냈을 때 그 공은 빨간색이 아니다. 그럼 그 공이 녹색일 확률은 ? 

우리가 찾아야하는 조건부 확률 : $P(G|\bar R) =\displaystyle \frac{P(G \cap \bar R)}{P(\bar R)}$

녹색공을 선택할 확률 : $P(G) =\displaystyle \frac{4}{20} = \frac{1}{5}$  

빨간공을 선택하지 않을 확률 : $P(\bar R) =\displaystyle \frac{12}{20} = \frac{3}{5}$ 

$P(G|\bar R) =\displaystyle \frac{P(G \cap \bar R)}{P(\bar R)} = \frac{P(G)}{P(\bar R)} =\frac{\frac{1}{5}}{\frac{3}{5}} = \frac{1}{3}$

# 1.7. 전체 확률의 정리 
- 표본 공간(sample space)는 배반 사건들의 집합으로 표현할 수 있음  
  $S$ : 표본공간(sample space)  
  $A_i$ : $A_1$, $A_2$ $\cdots$, $A_n$ 은 각각 배반 사건  

  즉 전체 표본 공간은 다음과 같은 식으로 배반 사건의 합으로 표현할 수 있음

  $$S = \sum_iA_i$$

- 표본 공간에 임의의 사건 K에 대한 확률은 총확률의 법칙을 바탕으로 다음과 같이 정리할 수 있음   

  $$P(K) = P(A_1 \cap K) + P(A_2 \cap K) + \cdots + P(A_n \cap K)$$  
  <br>

  $$P(K)  = \sum_{i=1}^nP(A_i \cap K)$$

  위 식에서 각각의 배반 사건에 대한 확률인 $P(A_i)$는 우리가 이미 알고 있는 사전 정보이므로 이 정보를 이용하여 조건부 확률 공식을 이용하여 총확률의 법칙을 정리할 수도 있음  

$$P(K|A) = \frac{P(A\cap K)}{P(A)}$$  
<br> 

$$P(A\cap K) = P(A)P(K|A)$$  
<br>  

$$P(K)  = \sum_{i=1}^nP(A_i)P(K | A_i)$$

예제) 전체 확률의 법칙  

A 제조사로부터 1000 개 집적회로 구입 : 불량확률 0.05  
B 제조사로부터 2000 개 집적회로 구입 : 불량확률 0.1   
C 제조사로부터 3000 개 집적회로 구입 : 불량확률 0.1

집적회로가 섞여있고 랜덤하게 집적회로를 선택했을 때 불량일 확률은 ? : P(D) 

선택한 집적회로가 A 일 경우 P(A) = $\displaystyle \frac{1000}{1000 + 2000 + 3000} = \frac{1}{6}$

선택한 집적회로가 B 일 경우 P(B) = $\displaystyle \frac{2000}{1000 + 2000 + 3000} = \frac{1}{3}$

선택한 집적회로가 C 일 경우 P(C) = $\displaystyle \frac{3000}{1000 + 2000 + 3000} = \frac{1}{2}$   

<br> 

A를 선택했을 때 불량일 확률 => P(D|A) = 0.05

B를 선택했을 때 불량일 확률 => P(D|B) = 0.1

C를 선택했을 때 불량일 확률 => P(D|C) = 0.1
<br> 

$P(D)$  
$= P(A)P(D|A) + P(B)P(D|B) + P(C)P(D|C)$  
$= \displaystyle \frac{1}{6}*0.05 + \frac{1}{3}*0.1 +\frac{1}{2}*0.1 = 0.09167$

# 1.8. 베이즈 정리  
- 앞서 다룬 조건부 확률은 사전 정보를 이용하여 사후 확률을 구하는 것을 정의한 것


- 지금 다루는 베이즈 정리는 사후 확률이 발생했을 경우에 사전 정보는 어떻게 변하는가에 관심을 둠


- 정리 자체는 단순하지만 많은 실생활에서 발생하는 문제 해결에 큰 도움을 주고 있음


- 다시 한번 앞서 다룬 조건부 확률은 다음과 같음.  

    - $P(A)$ : 각각의 배반 사건이며 이미 알고 있는 사전 정보 


$$P(B|A) = \frac{P(A \cap B)}{P(A)}$$      

<br>  

- 위 식을 교집합에 대해서 정리하면 다음과 같음    

    - $P(A \cap B)  = P(B|A)P(A) = P(A|B)P(B)$

    - $P(B) = P(A \cap B) + P(A^c \cap B)$

<br>   

- $P(B|A)$ 은 순서적으로 볼 때 대부분 사건 A가 먼저 발생하고 B가 발생하는 것   
    -  A는 원인, B는 결과의 형태를 가짐  

    -  원인의 가능성인 $P(A) 또는 P(A^c)$는 사건 B가 관측되기 이전의 확률
        - 사전 확률(prior probability)

 - 위 식을 아래 식에 대입하여 정리하면 바로 베이즈 정리임

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)} = \frac{P(B|A)P(A)}{P(A)P(B|A) + P(A^c)P(B | A^c )}$$

- 베이즈 정리는 특정 사건이 일어났을 때 사전 정보를 이용하여 쉽게 문제를 해결하는 방법을 제공  


- 원인과 결과 형태의 문제에서 결과에 대한 원인 분석 가능  


예제) 베이즈 정리   

결함 점검 장비   
  - 95% 확률로 실제로 결함이 있는걸 확인(TP)
  - 1% 확률로 실제로 결함이 없는데 구조적 결함이 있다고 보고(FP)  
  - 2% 확률로 실제 결함이 있음 (TP + FN)
실제 항공기의 2%가 결함이 있을때 항공기가 실제 결함이 있고 장비가 결함이 있다고 보고할 확률은 P(D|B) ?

$P(D)$ : 항공기 결함확률 (TP + FN)

$P(B)$ : 장비가 결함이 있다고 보고하는 확률 (TP + FP)

${P(D|B)}$

= $\displaystyle \frac{P(D \cap B)}{P(B)}$    

= $\displaystyle \frac{P(B |D )P(B)}{P(B)}$    

= $\displaystyle \frac{P(B |D )P(B)}{P(B|D)P(D) + P(B|\bar D)P(\bar D)}$    
       
= $\displaystyle \frac{0.95*0.02}{(0.95*0.02)+(0.01*0.98)}$   

= $0.66$

# 2.3. 확률 변수로 정의되는 사건  
- $X$를 확률 변수, $x$는 실수값일 때 확률 변수 $X$에 숫자가 할당되는 모든 사건을 $A_x$로 정의하면 다음과 같이 정리할 수 있음 

$$A_x = \{w | X(w) = x \} = [X=x]$$

- $A_x$는 하나의 사건이므로 확률을 가질 수도 있음

$$p=P[A_x]$$

- $P[X \leq x]$ : $X$가 $x$보다 작거나 같은 값을 가질 확률  
<br> 

- $P[X > x]$ : $X$가 $x$보다 큰 값을 가질 확률  = 1- $P[X \leq x]$  
<br> 

- $P[a < X < b]$ : $X$가 a와 b 사이의 값을 가질 확률 

<br>

$$S = \{HH, HT, TH, TT \}$$

$$\{0, 1, 2 \}$$  

앞면이 1번 이하로 나오는 사건에 대한에 대한 확률은 다음과 같이 구할수 있음

# 2.4. 누적 분포 함수  
- X 를 확률 변수 x 를 어떤 수라고 가정할 때 X에 대한 누적 분포 함수는 다음과 같이 정의됨 

$$F_x(x) = P[X \leq x], \quad ( -\infty < x < \infty)$$   

- $F_x(x)$는 확률 변수 X가 x보다 작거나 같을 가질 확률을 의미함  


- $F_x(x)$가 가지는 특징
    - 감소하지 않은 함수이다. 즉 이전값을 계속 누적하기 때문임  

    - $0 \leq F_x(x) \leq 1$  

    - $F_x(\infty) = 1$  

    - $F_x(-\infty) = 0$  

    - $P[a < X \leq b] = F_x(b) - F_x(a)$  

    - $P[X > a] = 1 - P[X \leq a] = 1 - F_x(a)$

  - 0 , $x < 0$  

  - $x + \displaystyle \frac{1}{2}, \quad 0 \leq x \leq \frac{1}{2}$    

  - 1, $\displaystyle \quad x > \frac{1}{2}$

# 3.1. 개요  
- 데이터 집합 $X_1, X_2, \cdots, X_n$이 주어지면 이에 대한 산술 평균은 다음과 같음  

$$\bar X = \frac{X_1  + X_2  + \cdots + X_n}{N}$$  


- 위의 수들이 서로 다른 빈도로 발생하면 가중 산술 평균은 다음과 같이 정리할 수 있음 

$$X_1 이 \quad k_1개$$ 
$$X_2 이 \quad k_2개$$  
$$X_3 이 \quad k_3개$$  
$$\vdots$$  
$$X_n 이 \quad k_n개$$  
<br>

$$N = k_1 + k_2 + \cdots + k_n$$    

<br>

$$\bar X = \frac{X_1 * k_1 + X_2 * k_2 + \cdots + X_n * k_n}{N}$$  

<br>

$$\bar X = \frac{k_1}{N}*X_1 + \frac{k_2}{N}*X_2 + \cdots + \frac{k_n}{N}*X_n$$  



- 평균은 데이터 집합의 대표값  


- 기대값은 확률 변수와 관계된 평균을 구하는 과정에서 사용되는 용어

    - 표본 평균 $\bar x$
        - $\displaystyle \bar x = 1*\frac{2}{5}+2*\frac{1}{5}+5*\frac{1}{5} + 6*\frac{1}{5}=3$ 

    - 관측된 값에 자료중 그 값이 차지하는 비율을 곱하여 값을 구함  

    - 표본 크기 n이 계속 커지면 표본은 모집단에 가까워지며 표본평균은 모평균($\mu$)에 가까워짐

# 3.3. 확률 변수의 모멘트와 분산  
- 표본 분산
    - 표본 크기 : n    

    - $n_i$ : $x_i$의 빈도수   

    - 표본 분산  
     $S^2 =\displaystyle \frac{1}{n-1}\sum_{i =1}^kn_i(x-\bar x)^2 = \frac{n}{n-1}\sum_{i =1}^k(x-\bar x)^2p_i, \quad p_i = \frac{n_i}{n}$      

    - n을 계속 크게 하면 
        - $p_i \Rightarrow f_(x)$  

        - $\bar x \Rightarrow \mu$    

        - $\displaystyle \frac{n}{n-1} = 1$   

        - 표본 분산 $\Rightarrow$ 모분산   <br> 

    - $S^2 : 표본 분산,  \quad \sigma^2 : 모분산$   

    - $S^2 =\displaystyle \frac{n}{n-1}\sum_{i =1}^k(x-\bar x)^2p_i \quad \Rightarrow \sigma^2 = \sum_{i =1}^k(x_i-\mu)^2f(x)$     


- 확률 변수 $X$ 의 n차 모멘트는 다음과 같이 정의함    


- $E[X^n]$  

    - 이산 확률 변수 : $\sum_{i}x_i^np_x(x)$ 

    - 연속 확률 변수 : $\int_{-\infty}^{\infty}x^nf_x(x)dx$  


- 여기에서 n 이 1이면 기대값을 의미 


- $\bar X$를 평균이라고 정의할 때 n이 2일 때 모멘트는 분산을 의미함 

    - $E[(X-\bar X)^2]$
        - 이산 확률 변수 : $\sum_{i}(x_i - \bar X)^2p_x(x)$ 

        - 연속 확률 변수 : $\int_{-\infty}^{\infty}(x -\bar X)^2f_x(x)dx$  



- 분산의 제곱근은 표준편차 $\sigma$라고 함  


- 분산은 확률 밀도 함수나 확률 질량함수가 평균을 중심으로 얼마나 분포하고 있는지를 알려주는 척도   


- 분산 $Var(X) = E[X^2] - \mu^2$    


- 분산 $Var(aX+b) = a^2Var(X)$  
    - 위치의 변화를 주는 절편 b는 분산에 영향을 주지 않음   



예제) 기대값과 분산  

확률 변수 X가 다음과 같은 확률 밀도 함수 $f_X(x)$일 때 기대값과 분산은 ? 

  - $\frac{1}{4}, \quad 2 \leq x \leq 6$  
  
  - $0, \quad otherwise$   

기대값 : $E[X] = \int_{-\infty}^{\infty}f_X(x)dx = \frac{1}{4}\int_2^6xdx = 4$  


$E[X^2] =\frac{1}{4}\int_2^6x^2dx = \frac{52}{3}$  

분산 : $Var[X] = E[X^2] - E[X]^2 =\displaystyle \frac{52}{3} - 4^2 = \frac{4}{3}$

# 3.4. 조건부 기대값  
- 어떤 특성을 가지는 모집단의 부분집합의 평균을 계산해야 할 경우가 발생할 수도 있음

- 사건 A가 발생했다는 전제하에 랜덤 변수 X의 조건부 기대값은 다음과 같음 
    
    - $E[X | A]$
        - 이산 확률 변수 : $\sum_{i}x_ip_{X|A}(x_i|A)$   
            - $p_{X|A}(x|A) = \displaystyle \frac{p_X(x)}{P(A)}$   
              
        - 연속 확률 변수 : $\int_{-\infty}^{\infty}xf_{X|A}(x|A)dx$  
            - $f_{X|A}(x|A) = \displaystyle \frac{f_X(x)}{P(A)}$


예제) 조건부 기대값 

$A = \{ X \leq 55 \}$ 이고 확률 밀도 함수 $f_X(x)$가 다음과 같을 때 조건부 기대값 $E[X|A]$은 ?    

  - $\displaystyle \frac{1}{20}, \quad 40 \leq x \leq 60$

  - $0, \quad otherwise$  

$E[X|X \leq 55] =\displaystyle \frac{\int_{40}^{55}xf_X(x)dx}{F_X(55)} = \frac{1}{20}\frac{\int_{40}^{55}xdx}{F_X(55)}$  

$F_X(55) =\displaystyle \int_{40}^{55}f_X(x)dx = \frac{3}{4}$

$E[X|X \leq 55] =\displaystyle \frac{1}{20}\frac{\int_{40}^{55}xdx}{F_X(55)} = 47.5$

    - 성공 확률이 $P(S) = p$ 이면 실패 확률은 $P(F) = 1-p$  

    - 각 실험은 서로 독립이며 실험 결과는 다른 실험의 결과에 영향을 끼치지 않음

    - $p_X(x) = p^x(1-p)^{1-x}, \quad x = 0, 1$  


- 기대값 $E[X]$
    - $E[X]$ = 0 * (1-p) + 1 * p = p


- 분산 $\sigma_X^2$

    - 우선 $E[X^2]$에 대한 값을 구해야 함
        - $E[X^2] = 0^2 * (1-p) + 1^2(p) = p$  
    
        - $E[X]$ = p    

    - $\sigma_X^2 = E[X^2] - {E[X]}^2 = p - p^2 = p(1-p)$

# 4.3. 이항 분포  
- 성공할 확률이 p인 베르누이 시행을 n 번 반복 시행할 때의 확률 분포를 말함   


- 이항 분포는 평균값을 중심으로 대칭이 되는 특징을 가졌음 


- 성공 횟수 확률 변수 X, 반복횟수 n, 베르누이 시행 성공확률 p라고 하면 다음과 같이 표현됨   

  $$X \sim Bin(n, p)$$  


- 반복 시행 횟수 n이 1일 경우 베르누이 시행과 동일하게 됨  


- 이항 분포 확률 질량 함수 
    - n과 p에 따라 확률이 달라짐 
        - 이항분포의 모양을 결정  

        - 분포의 특성을 완전히 결정하는 값 : 모수(parameter) 

    - n번 시행후 x번 성공할 이항 분포는 다음과 같이 정리됨  

$$P(X = x) = f(x;n,p)={n\choose x}p^x(1-p)^{n-x}, \quad(x = 0,1,2, \cdots, n)$$  

<br>  

$${n\choose x} = \frac{n!}{(n-x)!x!}$$  


- 이항 분포 기대값 : $E[X] = np$  


- 이항 분포 분산 : $Var[X] = np(1-p)$ 


- 같은 확률을 가진 X,Y에 대해서 서로 독립이면 다음이 성립    

    - X~B(m,p)  

    - Y~B(n,p)     

    - X+Y ~ B(m+n, p)    


예제) 이항 분포 1  

동전을 10번 던져후 앞면이 6번 나왔다면 5번 던져 앞면이 나올 횟수의 기대값은?

앞면이 나올 확률 : 0.6 
$E[X] = np = 5 * 0.6 = 3$

    - Y ~ B(6, 0.4)

# 4.4. 기하 분포    
- 기하 확률 변수는 첫번째 성공이 발생할 때까지의 베르누이 시행을 나타냄    


- 성공할 확률이 p인 베르누이 시행을 성공할 때가지 시행하는 경우 실패 시행 횟수의 분포  
    - 표본공간 : $\{S, FS, FFS, FFFS, \cdots \}$  


- 확률 변수 X를 첫번째 성공이 발생할 때까지의 베르누이 시행 회수라 한다면 그 이전까지의   
   x-1번째까지는 실패했다고 볼 수 있음  


- 기하 분포의 확률 질량 함수는 다음과 같음 

$$p_X(x) = p(1-p)^{x-1}, \quad (x = 1,2,3,\cdots)$$  


- X ~ Geo(p) 라고 표현함 


- $P[X \leq x-1]$ : x번째 실험 이전에 성공할 확률   


- $P[X \geq x]$ : x+1번째 실험 이후에 성공할 확률   
    - $P[X \leq x] =\displaystyle \sum_{k = 0}^xp(1-p)^k = \frac{p-p(1-p)^{x+1}}{1-(1-p)} = 1-(1-p)^{x+1}$
    - 1 - $P[X \leq x-1] = (1-p)^x$     


- 제 1항이 p 이고 공비가 (1-p)인 등비급수 형태를 가지고 있음   


- 기하 분포 기대값 : $E[X] =\displaystyle \frac{1}{p}$  


- 기하 분포 분산 : $Var[X] =\displaystyle \frac{1-p}{p^2}$ 

예제)기하 분포  

가방안에 파란공 6개, 빨간공 4개가 있을 때 복원추출 방식으로 빨간공이 나올때까지 공을 꺼낸다고 했을 때 정확히 5개공을 꺼냈을 때 실험이 끝날 확률은 ?   

$P(R) = 0.4$  

$P(B) = 0.6$  

따라서 네번 연속 실패후 다섯번째 빨간공이 나오는 기하분포이므로 

$P[X=5] = p(1-p)^4 = 0.4 * 0.6^4 = 0.05184$

# 4.5. 푸아송 분포  
- 발생 가능성이 희박한 사건이 임의의 구간에서 평균적으로 $\lambda$ 번 발생  
    - 두 개이상의 사건이 동시에 발생할 확률은 0에 가깝다  

    - 어떤 구간에서의 사건 발생 확률은 구간의 크기에 비례한다  

    - 단위 구간은 더 작은 단위구간으로 나눌수 있다    

    - 구간을 나누었을 때 각 구간의 발생 빈도는 서로 독립  

    - 구간의 위치와 관계없이 동일 길이의 구간에서의 평균 발생 빈도는 동일    


- X : 위의 상황에서 해당 사건이 일어날 횟수 
    - X ~ Pois($\lambda$) 


- 단위 시간안에 평균적으로 발생하는 횟수에 대한 기대값이 $\lambda$ 일때 사건 발생 횟수 X 에 대한    
   확률 질량 함수는 다음과 같음    

$$P(X = x) = \frac{\lambda^xe^{-\lambda}}{x!}, \quad (x=0, 1, \cdots)$$  


- 푸아송 분포 기대값 : $E[X] = \lambda$ 


- 푸아송 분포 분산 : $Var[E] = \lambda$

$$P[X = x] =\displaystyle \frac{6^x}{x!}e^-6$$

1시간동안 2개의 메시지가 도착할 확률은 ?    
  - $P[X = 2] =\displaystyle \frac{6^2}{2!}e^-6 = 18*e^-6 = 0.0446$

1시간동안 아무 메시지도 오지 않을 확률은 ?    
  - $P[X = 0] =\displaystyle \frac{6^0}{0!}e^-6 = e^-6 = 0.0024$  
<br> 

1시간동안 적어도 3개의 메시지가 도착할 확률은 ?      
  - $P[X \geq 3] = 1 - P[X < 3] = 1 - e^-6(\displaystyle \frac{6^0}{0!} + \frac{6^1}{1!} + \frac{6^2}{2!}) = 0.938$

# 4.6. 정규 분포  
- 좌우 대칭 종 모양 형태의 연속 확률 분포   


- 양 극단으로 갈수록 값은 작아지나 축에 닿지는 않음 


- 일반적으로 대다수의 자료들이 정규 분포를 따르고 있음  


- 평균과 표준편차를 가지고 정규 분포를 정의함  
    - 즉, 정규 분포의 모수는 평균과 분산이라는 의미

    - $\mu$와 $\sigma^2$에 따른 정규분포의 확률 밀도 함수
        - X ~ N$(\mu, \sigma^2)$ 


$$f(x)=\displaystyle \frac{x}{\sqrt[]{2\pi\sigma^2}}e^{\displaystyle -\frac{(x-\mu)^2}{2\sigma^2}}$$

# 5.2. 결합 누적 분포 함수  
- 하나의 표본 공간을 갖는 $X, Y$로 정의된 확률 변수를 고려  


- 같은 집단에 있는 학생들의 학점이나 신장에 대해서 다룰때 각각에 대해서 확률 변수 X, Y라고 설정하면   
   결합 누적 분포 함수는 다음과 같이 정의됨 

$$F_{XY}(x,y) = P[X \leq x, Y \leq y]$$  


- 두 확률 변수를 동시에 다루고 있는 이중 확률 변수 관련 분포 함수임  


- 한계(marginal) 누적 분포 함수
    - $F_X(x) = P[X \leq x]$  

    - $F_Y(y) = P[Y \leq y]$


- 확률 변수 X,Y의 결합 누적 분포 함수가 다음과 같이 표현될 때 서로 독립이라고 정의함  

$$F_{XY}(x,y) = F_X(x)F_Y(y)$$


- 결합 누적 분포 함수의 특징 
    - $0 \leq F_{XY}(x,y) \leq 1$  

    - $P[x_1 < X \leq x_2, Y \leq y] = F_{XY}(x_2, y) - F_{XY}(x_1,y)$

    - $P[ X < x, y_1 < Y \leq y_2] = F_{XY}(x, y_2) - F_{XY}(x,y_1)$  


- $F_X(x) = F_{XY}(x, \infty)$  


- $F_Y(y) = F_{XY}(\infty, y)$  

예제) 결합 누적 분포 함수    

결합 누적 분포 함수 $F_{XY}(x,y), \quad F_X(x), \quad, F_Y(y) \;$가 주어졌을때 확률 변수 X 가 a보다 크고   
Y 가 b보다 큰 경우의 결합 확률은 ?    

$P[X > a, Y > b] = P[(X > a) \cap (Y > b)]$    

위 식은 다음과 같이 정리할 수 있음   

$P[X > a, Y > b] = 1 - \bar{P[(X > a) \cap (Y > b)]}$    

다시 위 식은 다음과 같이 변경하여 정리할 수 있음   

$P[X > a, Y > b]$  
$= 1 - P[(X \leq a) \cup (Y \leq b)]$    
$= 1 - P[X \leq a] - P[Y \leq b]+ P[X \leq a, Y \leq b]$    
$= 1 - F_X(a) - F_Y(b)+ F_{XY}(a,b)$    

<br> 

# 5.3. 이산 확률 변수  
- 두 확률 변수 X,Y가 이산 확률 변수라면 결합 확률 질량 함수는 다음과 같음  

$$p_{XY}(x,y) = P[X=x, Y=y]$$

- 결합 확률 질량 함수의 특성   

    - $0 \leq p_{XY}(x,y) \leq 1$  

    - $\sum_x\sum_yp_{XY}(x,y) = 1$ 

    - $\sum_{x \leq a}\sum_{y \leq b}p_{XY}(x,y) = F_{XY}(a,b)$ 


- 한계(marginal) 확률 질량 함수

    - $p_X(x) = \sum_yp_{XY}(x,y) = P[X=x]$  

    - $p_Y(y) = \sum_xp_{XY}(x,y) = P[Y=y]$  


- 랜덤변수 X, Y가 독립이라면 다음과 같음 

    - $p_{XY}(x,y) = p_X(x)p_Y(y)$

# 5.4. 연속 랜덤 변수
- 두 랜덤 변수 X,Y가 연속 랜덤 변수라면 누적 결합 분포 함수는 다음과 같음   

$$F_{XY}(x,y) = \int_{-\infty}^x\int_{-\infty}^yf_{XY}(u, v)dvdu$$  

- 결합 확률 밀도 함수의 특성 

    - 모든 x, y에 대해서 $f_{XY}(x,y) \geq 0$ 

    - $\int_{-\infty}^{\infty}f_{XY}(x,y)dxdy =1$  

    - $P[x_1 < X \leq x_2, y_1 < Y \leq y_2] = \int_{y1}^{y2}\int_{x_1}^{x_2}f_{XY}(x,y)dxdy$  

    - 한계(marginal) 확률 밀도 함수  
      - $f_X(x) = \int_{-\infty}^{\infty}f_{XY}(x,y)dy$

      - $f_Y(y) = \int_{-\infty}^{\infty}f_{XY}(x,y)dx$  

    - 랜덤 변수 X,Y가 독립이라면 다음이 성립됨 
      - $f_{XY}(x,y) = f_X(x)f_Y(y)$

# 5.5. 공분산
- 일변수의 분포 정도를 나타내는 분산과는 별개  


- 2개의 변수중 하나의 값이 상승하는 경향을 보일때 다른 값도 상승하는 경향을 보이면 공분산은 양의 값을 보여줌  


- 2개의 변수중 하나의 값이 상승하는 경향을 보일때 다른값이 하강하는 경향을 보이면 공분산은 음의 값을 보여줌  


- 공분산은 두 변수가 어떤 상관 관계가 있는지 보여주는 것이며 인과 관계를 알 수 있는 것이 아님  


- 즉 두 개의 확률 변수가 서로 어떤 관계를 가지며 변화하는가를 보여주는 기준  


- 다음과 같은 식으로 표현됨  

    $Cov(X,Y) = E[(X-\mu_X)(Y - \mu_Y)$    
    $= E[XY-\mu_XY -X\mu_Y + \mu_X\mu_Y]$    
    $= E[XY] -\mu_XE[Y] -E[X]\mu_Y + E[\mu_X\mu_Y]$    
    $= E[XY] -\mu_X\mu_Y -\mu_X\mu_Y + \mu_X\mu_Y$    
    $= E[XY] -\mu_X\mu_Y$    

- $\mu_X$, $\mu_y$, $E[XY]$는 결합 확률 분포, 주변 확률 분포를 통해 구함  

- 랜덤 변수 X,Y가 독립이라면 공분산은 다음과 같은 특징을 가짐  

$$E[XY] = \mu_X\mu_Y$$  
<br> 

$$Cov(X,Y) = 0$$    

- $Cov(X,Y) = 0$이 항상 랜덤변수 X,Y가 독립이라는 것을 의미하지는 않음  

- $Cov(X,Y) = 0$은 두 랜덤 변수가 상관되지 않음을 의미 

<br> 

# 5.6. 상관계수 
- 공분산은 확률 변수 X,Y 값에 따라 변동폭이 심함


- 정규화를 통해 일정 크기 범위내에 존재하도록 함


- 공분산을 각 확률 변수의 표준편차의 곱으로 나누어 정규화시킴  


- 다음 식으로 상관계수를 표현함  

$$Corr(X,Y) = \frac{Cov(X,Y)}{\sigma_X\sigma_Y}$$  


- 상관 계수는 다음과 같은 특징을 가짐  

$$ -1 \leq Corr(X,Y) \leq 1$$   


- 상관계수 $Corr(X,Y)$는 두 랜덤변수중 한 변수의 관찰된 값에 기반을 두어 얻은 다른 한 변수의   
  예상값이 얼마나 잘 맞는가를 가늠하는 값을 제공함   


- 따라서 랜덤변수 X,Y사이의 관계가 선형방정식으로 표현될 경우 둘 사이의 상관관계를 알기 쉬워지게 됨


- 절대값 1 에 가까울수록 강한 상관관계

예제)공분산 및 상관계수 

랜덤 변수 X,Y의 결합 확률 밀도 함수 $f_{XY}(x,y)$는 다음과 같음

   - $0, otherwise$

랜덤 변수 X,Y 한계 확률 밀도 함수는 ?

랜덤 변수 X,Y의 공분산은 ?

$E[X] =\displaystyle \mu_X = \frac{0 + 0.2}{2} = 0.1$   

$E[Y] =\displaystyle \mu_Y = \frac{1}{5} = 0.2$

### 결정계수  

- SST = SSR + SSE  => $\displaystyle \sum{(y_i - {\bar y})}^2 = \sum{(\hat y_i - \bar y )}^2 + \sum{(y_i - \hat y_i)}^2$


- SST : 실제값 $y_i$들과  평균 $\bar y$들의 차이의 제곱합  


- SSR : 예측값과 실제값 $y_i$들의 평균과의 차이의 제곱합  


- SSE : 예측값과 실제값 $y_i$들의 차이의 제곱합  



$R^2 =\displaystyle \frac{SSR}{SST} = 1 - \frac{SSE}{SST} = \frac{\alpha \sum{y_i} + \beta \sum{x_iy_i} - n{\bar y}^2}{\sum{y_i}^2 - n{\bar y}^2}$


- 결정계수는 0부터 1까지의 값을 가짐. 


- 회귀선이 모든 자료에 적합하면 SSE = 0, $R^2 = 1$이 됨  


- $R^2$값이 1에 가까울수록 종속변수를 정확하게 예측할 가능성이 높음 


- SST = 1257.33, SSE = 31.4348 => SSR = SST - SSE = 1225.895. 


- $R^2 =\displaystyle \frac{SSR}{SST} = \frac{1225.895}{1257.33} = 0.975$    


- 차량 가격이 판매액에 97.5%의 매우 큰 영향을 끼치고 있다고 판단할 수 있습니다.

- 평균 $\mu$, 분산 $\sigma^2$인 분포에서 n개의 확률 표본을 추출했을 경우 표본 평균 $\bar X$의 분포는 ?
    - 정규 분포인 경우
        - $X_1$ ~ N$(\mu, \sigma^2)$, $X_2$ ~ N$(\mu, \sigma^2)$ 이고 $X_1, X_2$가 독립이면 
            - $X_1 \pm X_2$ ~ N$(\mu_1 \pm \mu_2, \sigma_1^2 + \sigma_2^2)$    

        - $X_1, \cdots, X_n$ ~ N$(\mu, \sigma^2)$   

            - $X_1 + \cdots + X_n$ ~ N$(n\mu, n\sigma^2)$  

            - $\displaystyle \bar X  = \frac{X_1 + \cdots + X_n}{n}$ ~ N$(\mu, \sigma^2/n)$

# 6.3. 표집 분포    
-  모평균이 $\mu$ 이고 분산이 $\sigma^2$ 인 모집단에서 크기가 n 인   표본 $X_1$, $X_2$, $X_3$, ... $X_n$을   
   추출하였을 때 표본 평균과 분산은 다음과 같음    

    -  표본평균 $E[\bar{X}]$ = $\mu$ (모집단의 평균과 동일)   

    -   표본분산 $Var[\bar{X}]$ = $\frac{\sigma^2}{n}$ 

        - 정규 분포를 갖고 있는 모집단의 표본평균 $\bar{X}$는 정규 분포를 따름  

        - 모집단이 정규 분포가 아닌 경우에도 표본의 크기 n이 큰 경우(n > 30)이면 근사적으로 정규 분포를 따름  

        - 표본 평균을 정규화하면 다음과 같음  

$$Z = \frac{\bar{X} -\mu}{\frac{\sigma}{\sqrt[]{n}}} \sim N(0, 1)$$     

<br>


# 6.4. 중심 극한 정리      
- 대수의 법칙(law of large numbers)  
    - $X_1, X_2, \cdots, X_n$ 평균 $\mu$, 분산 $\sigma^2$인 모집단에서 추출된 확률 표본   
        - 표본 평균 : $\bar X =\displaystyle \frac{1}{n}\sum_{i = 1}^nX_i$  

        - $E[\bar X] = \mu, \quad Var[\bar X] = \sigma^2/n$  

        - n을 계속 크게 만들면 분산 $Var[\bar X] = \sigma^2/n \Rightarrow 0$  

        - $\bar X$는 $\mu$ 로 수렴함  


- 중심 극한 정리 

    - $X_1, X_2, \cdots, X_n$ 평균 $\mu$, 분산 $\sigma^2$인 모집단에서 추출된 확률 표본   

    - n이 커질수록 모집단의 형태와 관계없이 $\bar X$의 분포는 정규분포에 근사
